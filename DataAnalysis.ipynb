```python
import pandas as pd
import sqlite3 as sql
import numpy as np
from datetime import datetime
import plotly.graph_objects as go
import pandas_ta as ta
import matplotlib
import requests
from random import randint
```

```python
all_df = pd.read_sql('SELECT *, datetime(date,"localtime") FROM inv_hist_quotes', con=sql.connect('mydb.db'))
# all_df['date'] = all_df['date'].apply(lambda x : datetime.strptime(x, '%d-%m-%Y'))
all_df['date']
```

```python
all_df['name'].unique()
```

```python
dji_df = all_df[all_df['name'] == 'Dow Jones Industrial Average']
dji_df.sort_values(by=['date'], ascending=False)
dji_df_sl = dji_df[dji_df['date'].between(datetime(2018,4,14), datetime(2022,4,14))]
```

```python
fig = go.Figure(data=[ 
    go.Scatter(
        x=dji_df_sl['date'], 
        y=dji_df_sl['close'],
        line=dict(
            color='green', 
            width=1)
        ),
    go.Scatter(
        x=dji_df_sl['date'], 
        y=dji_df_sl['close'],
        line=dict(
            color='red', 
            width=1)
        )
    ])

fig.add_hline(y=dji_df_sl['close'].std())

fig.show()   
```
